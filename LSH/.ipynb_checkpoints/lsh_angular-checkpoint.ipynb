# 7.4 Minhash para la distancia angular
En esta primera parte definiremos la probabilidad en funcion a la distancia, cantidad de funciones hash (**r**) y cantidad de tablas de hash (**b**)

```python
def p(d, r=1, b=1):
    return 1 - (1 - (1 - d / pi) ** r) ** b
```

Probamos

```python
print(p(pi))
print(p(pi/2))
print(p(0))
print(p(1))
```

## Graficamos p en funcion de r y b

```python
def plot_prob(x, r=1, b=1):
    y = p(x, r, b)
    plt.plot(x, y)
    plt.title("Probabilidad de colision para r=%d y b=%d" % (r, b))
    plt.xlabel("Distancia angular")
    plt.ylabel("Probabilidad")
```

Distancia de 0 a pi, en rangos de .01

```python
x = np.arange(0, pi, .01)
```

Con r=1 y b=1

```python
plot_prob(x)
plt.show()
```

Aumentamos r

```python
for i in range(1, 11):
    plot_prob(x, r=i)

plt.show()
```

Aumentamos b

```python
for i in range(1, 11):
    plot_prob(x, b=i)

plt.show()
```

Ahora buscaremos valores de **r** y **b** para las probabilidades de colision:
* 0.95 para distancia igual o menor a 0,175082925 (10 grados)
* 0.05 para distancias igual o mayores a 0,785398163 (45 grados)

```python
r=20
b=10
print(p(10 * pi / 180, r=r, b=b))
print(p(45 * pi / 180, r=r, b=b))
```

Graficamos la probabilidad en funcion a la distancia con estos valores de **r** y **b**

```python
plot_prob(x, r=r, b=b)
plt.show()
```

# 7.4.1 El mtodo de los hiperplanos

Funcion auxiliar, transformar array de bits en numero entero.

```python
# https://stackoverflow.com/questions/12461361/bits-list-to-integer-in-python#answer-12461400
def bitArray(bitlist):
    out = 0
    for bit in bitlist:
        out = (out << 1) | bit
    return out
```

Funcion para crear x hiperplanos de d dimensiones con valores -1 y 1.

```python
def generar_hiperplanos(num_dim, num_vect):
    return np.array([[1 if np.random.randint(2) else -1 for x in range(num_dim)] for y in range(num_vect)])
```

Funcion para crear x hiperplanos de d dimensiones sin restringir sus valores como la funcion anterior.

```python
def generar_hiperplanos2(num_dim, num_vect):
    return np.random.rand(num_vect, num_dim) - .5
```

Ejemplo:

```python
hiperplanos = generar_hiperplanos(2, 5)
hiperplanos
```

Creamos un vector y calculamos su producto interno con cada hiperplano.

```python
vector = np.array([10, 5])
vector
```

```python
for h in hiperplanos:
    print(vector[0] * h[0] + vector[1] * h[1])
```

Podemos hacerlo de una forma mas optima:

```python
vector.dot(hiperplanos.transpose())
```

Cada producto interno sera un minhash, su valor es 1 si el producto es igual o mayor a 0, de lo contrario sera 0.

```python
minhashes = vector.dot(hiperplanos.transpose()) >= 0
minhashes
```

Ahora interpretaremos cada minhash como si fueran bits y los traduciremos a un numero entero:

```python
bitArray(minhashes)
```

## Generar r*b minhashes para b tablas de hash
Dado lo anterior, crearemos una funcion que nos devuelva un entero a partir de un vector y N hiperplanos. Mas adelanta utilizaremos este entero como hash para guardar los vectores en las **b** tablas de hash.

```python
def hash_angular(vector, hiperplanos):
    return bitArray(vector.dot(hiperplanos.transpose()) >= 0)
```

Probamos con el vector y los hiperplanos anteriores, deberia devolver el mismo valor.

```python
hash_angular(vector, hiperplanos)
```

Ahora generaremos **r** * **b** hiperplanos para crear un hash (formado por **r** minhash) por cada tabla de hash (**b**)

```python
hiperplanos = [generar_hiperplanos(2, r) for i in range(b)]
print("(# de hiperplanos, # de minhashes, dimensiones)")
np.shape(hiperplanos)
```

Por ejemplo los **b** hashes para guardar el vector de ejemplo en las **b** tablas de hash seran:

```python
[hash_angular(vector, hiperplanos[j]) for j in range(b)]
```

# Caso de uso
Generaremos 1000 vectores de dos dimensiones con valores aleatorios. Utilizaremos la funcion **hash_angular** para generar **b** hashes por vector y asi guardarlo en las **b** tablas de hash.

Luego, dado un nuevo vector, calcularemos sus hashes y mostraremos sus candidatos a vecinos.

```python
def lsh_test(r, b, hiperplanos, vectores, vector):
    
    # Tablas de hash.
    tables = [{} for i in range(b)]

    # Guardar cada vector en su correspondente bucket de cada tabla hash.
    for item in vectores:
        for i in range(b):
            hash_val = hash_angular(item, hiperplanos[i])

            if hash_val not in tables[i]:
                tables[i][hash_val] = []

            tables[i][hash_val].append(item)

    # Calcular vecinos del vector pasado por parametro.
    vecinos = None
    for i in range(b):
        hash_val = hash_angular(vector, hiperplanos[i])
        if hash_val in tables[i]:
            if vecinos is None:
                vecinos = tables[i][hash_val]
            
            # Probar con un dataframe, ya que con esta solucion se repiten muchos vectores.
            vecinos = np.concatenate((vecinos, tables[i][hash_val]), axis=0)
    
    # Graficamos
    plt.figure(figsize=(10, 10))
    plt.scatter(vectores[:, 0], vectores[:, 1], label="Vectores")
    plt.scatter(vecinos[:, 0], vecinos[:, 1], color='gold', label="Candidatos a vecinos")
    plt.plot(vector[0], vector[1], 'r^', markersize=10, label="Vector objetivo")
    plt.title("Vectores")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.xlim(min(vectores[:,0]), max(vectores[:,0]))
    plt.ylim(min(vectores[:,1]), max(vectores[:,1]))
    plt.legend(loc=1)
    plt.show()
```

Generamos los vectores, el vector y nuevamente los hiperplanos

```python
hiperplanos = [generar_hiperplanos(2, r) for i in range(b)]
vectores = np.random.rand(1000, 2)
vector = np.array([.2, .3])
```

```python
lsh_test(r, b, hiperplanos, vectores, vector)
```

Aproximadamente la mitad de los vectores son candidatos a vecinos, probaremos normializando todos los vectores.

```python
mean = np.mean(vectores)
std = np.std(vectores)

vectores_n = (vectores - mean) / std
vector_n = (vector - mean) / std
```

```python
lsh_test(r, b, hiperplanos, vectores_n, vector_n)
```

Ahora solo un cuarto de los vectores son candidatos a vecinos.

Observamos una relacion entre el angulo que forman los candidatos a vecinos y la limitacion de los hiperplanos a valores de -1 y 1.

Probaremos generando nuevos hiperplanos sin esta limitacion.

```python
hiperplanos2 = [generar_hiperplanos2(2, r) for i in range(b)]
```

```python
lsh_test(r, b, hiperplanos2, vectores_n, vector_n)
```

Mucho mejor, ahora el angulo entre los candidatos a vecinos si parece estar mas cercano a los valores que estimamos con **r** y **b**.

Podemos observar la influencia utilizar la distancia angular sobre los vectores que son elejidos como candidatos a vecinos.

Que pasara si nos acercamos al centro?

```python
vector_centrico = np.array([-.1, .2])
lsh_test(r, b, hiperplanos2, vectores_n, vector_centrico)
```

* El angulo formado entre los candidatos a vecinos ha cambiado, pareceria haber una relacion con la posicion del vector.
* La distancia angular pareceria ser un buen sustituto a la distancia euclidiana para observaciones distantes al centro de nuestra muestra/poblacion, pero no para valores centricos.

Probaremos con valores de **r** y **b** en uno.

```python
r2=1
b2=1
print(p(10 * pi / 180, r=r2, b=b2))
print(p(45 * pi / 180, r=r2, b=b2))
```
