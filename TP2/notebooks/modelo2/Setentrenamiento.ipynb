#### Postulantes por nivel de educación

```python
#Se arma el frame de postulantes por nivel de educacion
df_post_edu=df_post_edu.append([df_post_edu_h_15_4,df_post_edu_d_15_4])
del df_post_edu_d_15_4
del df_post_edu_h_15_4
```

Como muchos declararon más de un nivel de educación, se filtrará el nivel más alto como hecho en el TP1, de acuerdo a las siguintes categorías ordenadas (modifiqué para que los puntajes queden como los de Sebas y Jose)

Secundario - Otro - Terciario - Universitario - Posgrado - Master - Doctorado

y dentro de cada categoría el orden es:

abandonado - en curso - graduado

```python
#Como no hay una única declaración de estudios, se puntuará los estudios declarados por nivel del curso y estado del curso
df_curso=pd.DataFrame({'nombre':pd.Categorical(['Secundario','Otro','Terciario/Técnico','Universitario','Posgrado','Master','Doctorado']),'puntos':[0,3,6,9,12,15,18]})
df_estado=pd.DataFrame({'estado':pd.Categorical(['Abandonado','En Curso','Graduado']),'puntos2':[0,1,2]})
```

```python
#se juntan las tablas de postulaciones y valor del nombre del curso
df_post_edu_unico=pd.merge(df_post_edu,df_curso,on='nombre')
```

```python
#y valor del estado del curso
df_post_edu_unico=pd.merge(df_post_edu_unico,df_estado,on='estado')
```

```python
#se cuantifica el curso con un valor preestablecido
df_post_edu_unico['valor_educ']=df_post_edu_unico.puntos+df_post_edu_unico.puntos2
```

```python
#se ordena por valor del curso
df_post_edu_unico=df_post_edu_unico.iloc[:,[0,1,2,5]].sort_values(by='valor_educ',ascending=False)
```

```python
#elimina a los repetidos posteriores (o sea, se queda con el primero que es el del valor más alto)
df_post_edu_unico=df_post_edu_unico.drop_duplicates('idpostulante')
```

```python
del df_curso
del df_estado
del df_post_edu
```

```python
df_post_edu_unico.iloc[:,0:4].sort_values(by='idpostulante').sample(10)
```

#### Gráfico de barras de la distribución por estudios

```python
#define una tabla por curso y estado del curso
df=pd.DataFrame({'curso':df_post_edu_unico.nombre,'estado':df_post_edu_unico.estado,'q':1})
```

```python
df=pd.crosstab(df_post_edu_unico.nombre,df_post_edu_unico.estado)
    #crea una columna de total de postulantes por curso con el objetivo de ordenar del curso más estudiado 
    #al menos estudiado en orden decreciente
df['total']=df.T.sum()
df=df.sort_values(by='total',ascending=False)
    #elimina la columna del total de avisos
df=df.iloc[:,0:3] 
```

```python
#grafica las barras empiladas por curso y estado
plt.figure()
df.plot.bar(stacked=True,figsize=(12,6))
plt.title('cantidad de postulantes por mayor nivel de estudios declarado discriminado por el estado del curso')
plt.xlabel('nivel de estudio de los postulantes',fontsize=12)
plt.ylabel('cantidad de postulantes',fontsize=12)
```

Se juntan las tablas de todos los postulantes en una tabla única con la educación, el género y la edad

### Depuración de los datos de entrenamiento

Se depurarán los datos de entrenamiento de ruído, considerando este como datos imposibles, mal introducidos, etc. Entre ellos se eliminarán las edades menor que 15 y mayor a 75. También a los postulantes a cargos de Gerencia/altagerencia/dirección a los que tienen nivel educativo menor a 8 (técnico graduado) y a los que se candidatan a primer empleo con más de 50 años.

```python
df_postulaciones=df_postulaciones[(df_postulaciones.edad>=15)&(df_postulaciones.edad<=75)]
df_postulaciones=df_postulaciones[~((df_postulaciones.nivel_laboral=='Gerencia / Alta Gerencia / Dirección')&(df_postulaciones.valor_educ<8))]
df_postulaciones=df_postulaciones[~((df_postulaciones.tipo_de_trabajo=='Primer empleo')&(df_postulaciones.edad>50))]
```

```python
#número final de postulaciones "válidas", postulantes y avisos con detalles
print("postulaciones válidas", len(df_postulaciones))
print("avisos con detalles", len(df_avisos_detalle))
print("número de postulantes", len(df_postulante))
```

## Casos negativos

#### Agregado de datos en el frame de postulaciones con el fin de entrenar el algoritmo de ML. Estos datos corresponden a postulantes reales y avisos reales pero que NO están en el set de postulaciones. Se agrega una columna post que indica si esa postulación es real (1) o es un dato agregado (0). La idea es que el algoritmo pueda separar estos dos sets (mediante un hiperplano?)

```python
#se agrega la columna post
df_postulaciones['post']=[1]*len(df_postulaciones)
```

```python
#numero de datos agregados
nag=3000000

#id aleatorios de postulantes y avisos
idnumpost=np.random.choice(df_postulante.idpostulante.unique(), nag)
idnumadvice= np.random.choice(df_avisos_detalle.idaviso.unique(), nag)

#creo un data frame con pares de datos adicionales inventados (aunque de postulantes reales y avisos reales)
df3=[]
for i in range(nag):
    df3.append({
        'idaviso':idnumadvice[i],
        'idpostulante':idnumpost[i]})
df3 = pd.DataFrame(df3, columns=['idaviso', 'idpostulante'])
```

```python
#completo el data frame com los datos que faltan de aviso y postulante
df3 = pd.merge(df3, df_postulante, on='idpostulante')
df3 = pd.merge(df3, df_avisos_detalle.iloc[:,[0,2,7,8,9,10]], on='idaviso')

#agrego la columna de post con 0 porque no son postulaciones reales
df3['post']=0
```

```python
#ordena las columnas
df3=df3.iloc[:,[0,8,9,10,11,12,1,4,6,7,13]]
```

```python
#lo adiciono al frame de postulaciones y elimino las posibles línes repetidas (como borra las posteriores, es seguro que las 
#que originalmente estaban en el frame de postulacioens seguirán, se borrarán los agregados repetidos)
df_postulaciones=df_postulaciones.append(df3,ignore_index=True).drop_duplicates(['idaviso','idpostulante']) 
```

```python
del df3
```
